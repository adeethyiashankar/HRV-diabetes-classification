Move images from Diabetes and Healthy to All

```python
def moveImages_Diabetes(diabetes_dir, all_dir):
  for i in range(9):
    for files in os.listdir(diabetes_dir+f'{(i+1):03d}'+'/'):
      if files.endswith('.csv'):
        shutil.copy(diabetes_dir+f'{(i+1):03d}'+'/'+files, all_dir+'d'+f'{(i+1):03d}'+'_'+files)
```

```python
def moveImages_Healthy(healthy_dir, all_dir):
  for i in range(20):
    if i != 11:
      for files in os.listdir(healthy_dir+f'{(i+1):03d}'+'/'):
        if files.endswith('.csv'):
          shutil.copy(healthy_dir+f'{(i+1):03d}'+'/'+files, all_dir+'h'+f'{(i+1):03d}'+'_'+files)
```

Check how many images are in each folder

```python
def numImages(all_dir, healthy_dir, diabetes_dir):
  total_files = 0
  for files in os.listdir(all_dir):
    total_files += 1
  print(total_files)

  healthy = 0
  for a, b, files in os.walk(healthy_dir):
    for name in files:
      healthy += 1
  print(healthy)

  diabetes = 0
  for a, b, files in os.walk(diabetes_dir):
    for name in files:
      diabetes += 1
  print(diabetes)
```

```python
diabetes_dir = 'gdrive/My Drive/Summer Research/HRV/Outlier Free/Diabetes/'
healthy_dir = 'gdrive/My Drive/Summer Research/HRV/Outlier Free/Healthy/'
all_dir = 'gdrive/My Drive/Summer Research/HRV/Outlier Free/All/'

moveImages_Diabetes(diabetes_dir, all_dir)
moveImages_Healthy(healthy_dir, all_dir)
numImages(all_dir, healthy_dir, diabetes_dir)
```
